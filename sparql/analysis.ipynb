```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("query1.csv")
df.head()
```

```python
fig, axs = plt.subplots(3,1, figsize=(14,10))

ax = sns.lineplot(x=df.index, y=df.releases_count, ax=axs[0])
ax.set_xlabel('Countries ranked from most releases to least')
ax.set_ylabel('# releases')
ax.set_title('Number of releases by countries',fontdict={'fontsize':20})

ax = sns.lineplot(x=df.index, y=df.happiness_score, ax=axs[1], color='grey')
ax = sns.lineplot(x=df.index, y=df.happiness_score.rolling(7).mean(), ax=axs[1], linewidth=2.5)
ax.set_xlabel('Countries ranked from most releases to least')
ax.set_ylabel('Happiness Score')
ax.set_title('Happiness Score by countries',fontdict={'fontsize':20})

ax = sns.lineplot(x=df.index, y=df.gdp_2019, ax=axs[2], color='grey')
ax = sns.lineplot(x=df.index, y=df.gdp_2019.rolling(7).mean(), ax=axs[2], linewidth=2.5)
ax.set_xlabel('Countries ranked from most releases to least')
ax.set_ylabel('GDP per capita')
ax.set_title('GDP per capita by countries',fontdict={'fontsize':20})

plt.tight_layout()
plt.savefig("query1.png", format='png', dpi=150)
```

```python
sns.set(font_scale=1.2)
plt.figure(figsize=(9, 6))
corr = df.corr()
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(corr, dtype=bool))
heatmap = sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':20});
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig("query1_corr.png", format='png', dpi=150)
```
